### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total = tienda['Precio'].sum()
ingreso_total2 = tienda2['Precio'].sum()
ingreso_total3 = tienda3['Precio'].sum()
ingreso_total4 = tienda4['Precio'].sum()
```

```python
print(f'Los ingresos de cada tienda son: \nPara la tienda 1: {ingreso_total}$.\nPara la tienda 2: {ingreso_total2}$.\nPara la tienda 3: {ingreso_total3}$.\nPara la tienda 4: {ingreso_total4}$.')
```

# 2. Ventas por categoría

```python
tienda['Categoría del Producto'].unique()
```

```python
venta_categoria_tienda = tienda.groupby('Categoría del Producto')[['Categoría del Producto']].value_counts().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas')
venta_categoria_tienda.plot(kind='barh', figsize=(10,6), color='red', x='Categoría del Producto', y='Cantidad de Ventas')
plt.xlabel('Cantidad de Ventas por Categoría')
plt.title('Cantidad de Ventas de la Tienda 1')
```

```python
venta_categoria_tienda2 = tienda2.groupby('Categoría del Producto')[['Categoría del Producto']].value_counts().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas')
venta_categoria_tienda2.plot(kind='barh', figsize=(10,6), color='yellow', x='Categoría del Producto', y='Cantidad de Ventas')
plt.xlabel('Cantidad de Ventas por Categoría')
plt.title('Cantidad de Ventas de la Tienda 2')
```

```python
venta_categoria_tienda3 = tienda3.groupby('Categoría del Producto')[['Categoría del Producto']].value_counts().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas')
venta_categoria_tienda3.plot(kind='barh', figsize=(10,6), color='Blue', x='Categoría del Producto', y='Cantidad de Ventas')
plt.xlabel('Cantidad de Ventas por Categoría')
plt.title('Cantidad de Ventas de la Tienda 3')
```

```python
venta_categoria_tienda4 = tienda4.groupby('Categoría del Producto')[['Categoría del Producto']].value_counts().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas')
venta_categoria_tienda4.plot(kind='barh', figsize=(10,6), color='pink', x='Categoría del Producto', y='Cantidad de Ventas')
plt.xlabel('Cantidad de Ventas por Categoría')
plt.title('Cantidad de Ventas de la Tienda 4')
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda = tienda.Calificación.mean().round(2)
promedio_tienda2 = tienda2.Calificación.mean().round(2)
promedio_tienda3 = tienda3.Calificación.mean().round(2)
promedio_tienda4 = tienda4.Calificación.mean().round(2)
```

```python
print(f'Calificacion promedio de cada tienda:\nTienda 1: {promedio_tienda}.\nTienda 2: {promedio_tienda2}.\nTienda 3: {promedio_tienda3}.\nTienda 4: {promedio_tienda4}.')
```
