### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total1 = tienda['Precio'].sum()
ingreso_total2 = tienda2['Precio'].sum()
ingreso_total3 = tienda3['Precio'].sum()
ingreso_total4 = tienda4['Precio'].sum()
total = ingreso_total1 + ingreso_total2 + ingreso_total3 + ingreso_total4
```

```python
print(f'Los ingresos de cada tienda son: \nPara la tienda 1: {ingreso_total1}$.\nPara la tienda 2: {ingreso_total2}$.\nPara la tienda 3: {ingreso_total3}$.\nPara la tienda 4: {ingreso_total4}$.\nEl total de ingresos es de: {total}$')
```

```python
ingresos = [ingreso_total1, ingreso_total2, ingreso_total3, ingreso_total4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.pie(ingresos, labels=nombres_tiendas, autopct='%1.1f%%')
plt.show()
```

Observamos:

El desempeño de las 4 tiendas es equitativo, el negocio es estable.

Hay un mínimo liderazgo de la Tienda 1 en cuanto a ingresos con 26.1%, pero la diferencia con la última de la lista (la Tienda 4 con 23.6%) es marginal, de apenas un 2.5%.

```python
def concatenar(lista):
  string = ''
  if len(lista)==1:
    string = lista[0]
  for i in range(len(lista)):
    if i==0:
      string = lista[i]
    else:
      string = string + ' - ' + lista[i]
  return string
```

```python
conteo_productos = tienda.Producto.value_counts()
mas_vendido = 'Tienda 1: ' + concatenar((conteo_productos[conteo_productos==tienda.Producto.value_counts().max()]).index.tolist())
menos_vendido = 'Tienda 1: ' + concatenar((conteo_productos[conteo_productos==tienda.Producto.value_counts().min()]).index.tolist())
```

```python
conteo_productos2 = tienda2.Producto.value_counts()
mas_vendido2 = 'Tienda 2: ' + concatenar((conteo_productos2[conteo_productos2==tienda2.Producto.value_counts().max()]).index.tolist())
menos_vendido2 = 'Tienda 2: ' + concatenar((conteo_productos2[conteo_productos2==tienda2.Producto.value_counts().min()]).index.tolist())
```

```python
conteo_productos3 = tienda3.Producto.value_counts()
mas_vendido3 = 'Tienda 3: ' + concatenar((conteo_productos3[conteo_productos3==tienda3.Producto.value_counts().max()]).index.tolist())
menos_vendido3 = 'Tienda 3: ' + concatenar((conteo_productos3[conteo_productos3==tienda3.Producto.value_counts().min()]).index.tolist())
```

```python
conteo_productos4 = tienda4.Producto.value_counts()
mas_vendido4 = 'Tienda 4: ' + concatenar((conteo_productos4[conteo_productos4==tienda4.Producto.value_counts().max()]).index.tolist())
menos_vendido4 = 'Tienda 4: ' + concatenar((conteo_productos4[conteo_productos4==tienda4.Producto.value_counts().min()]).index.tolist())
```

```python
mas_vendidos_totales = {'Tiendas y sus productos': [mas_vendido, mas_vendido2, mas_vendido3, mas_vendido4], 'Cantidad de ventas': [conteo_productos.max(),conteo_productos2.max(),conteo_productos3.max(),conteo_productos4.max()]}
df_mas_vendidos_totales = pd.DataFrame(mas_vendidos_totales)
df_mas_vendidos_totales.plot(kind='bar', y='Cantidad de ventas', x='Tiendas y sus productos', figsize=(8,4), color='green')
```

# 5. Envío promedio por tienda

```python
promedio_envio = tienda['Costo de envío'].mean().round(2)
promedio_envio2 = tienda2['Costo de envío'].mean().round(2)
promedio_envio3 = tienda3['Costo de envío'].mean().round(2)
promedio_envio4 = tienda4['Costo de envío'].mean().round(2)
```

```python
print(f'Promedio de envio por tienda:\nTienda 1: {promedio_envio}$.\nTienda 2: {promedio_envio2}$.\nTienda 3: {promedio_envio3}$.\nTienda 4: {promedio_envio4}$.\n')
```

```python
df_envios = pd.DataFrame({
    'Tienda 1': tienda['Costo de envío'],
    'Tienda 2': tienda2['Costo de envío'],
    'Tienda 3': tienda3['Costo de envío'],
    'Tienda 4': tienda4['Costo de envío']
})
df_envios.boxplot(figsize=(8, 6))
```

Observamos:
- Las 4 Tiendas tienen un comportamiento de precios prácticamente idéntico. Ninguna tienda es significativamente más barata o más cara que las otras en cuanto a envíos.

- Existen muchísimos envíos excepcionalmente caros (outliers) que cuestan entre 80,000 y 160,000. Esto podría ser porque están enviando productos muy pesados o enviando a ciudades muy lejanas.

- El promedio es engañoso, los valores atipicos estan desplazando el promedio, aunque la realidad es que la mitad de los clientes pagan menos de 15,000.

# Conclusión

Con base en los datos expuestos, la recomendación estratégica para el Sr. Juan es vender la Tienda 4. Las principales razones son:

- Menor Retorno de Inversión (ROI): Al ser la sucursal con el menor porcentaje de ingresos (23.6%), retiene capital que podría estar rindiendo mejor en otras áreas del negocio.

- Falta de un "Producto Estrella" dominante: A diferencia de la Tienda 2, que logra picos de hasta 65 ventas en un solo artículo, el catálogo de la Tienda 4 no genera el mismo nivel de tracción masiva.

- Eficiencia Logística: Como todas las tiendas sufren por igual los costos de envío extremadamente altos, cerrar la sucursal que menos dinero ingresa ayuda a aliviar inmediatamente el impacto de esos costos atípicos en el balance general.

Vender la Tienda 4 le permitirá al Sr. Juan liquidar un activo de menor rendimiento y redirigir ese capital operativo para inyectar inventario en la Tienda 1 (la de mayor facturación) y potenciar estrategias de marketing para los productos líderes de la Tienda 2.
